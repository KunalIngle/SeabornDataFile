### Seaborn
Seaborn is a visualization library for making statistical graphics in Python that is more adapted for efficient use with the Pandas data frames. Seaborn is built on top of matplotlib.
The library provides a lot of flexibility when it comes to plotting from data frames allowing users to choose from a wide range of plotting styles while mapping the set of features from the data efficiently.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

https://www.kaggle.com/ranjeetjain3/seaborn-tips-dataset?select=tips.csv

The tips data set is a simple dataset that consists of observations on tip providers in restaurants. The data consists of the following features :

**total_bill: The total bill paid by the customer.**

**tip: The tip provided by the customer.**

**sex: The gender of the customer.**

**smoker: If the customer is a smoker or not.**

**day: The day of the week when the observation was made.**

**time: The time of the observation, whether at lunch or dinner etc.**

**size: The size of the group whether there were multiple members.**

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Distplots allows us to show the distribution of the univariate set of observations. Univariate means just one variable. It  plots the occurrences or density of the specified feature in the dataset
### A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin

```python
sns.distplot(tips['total_bill']) # Here, line is KDE - Kernel density estimate, output is Histogram.
```

```python
#plt.figure(figsize=(10,8))
sns.distplot(tips['total_bill'], kde=False)
#plt.xlim(10,40)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

### Jointplot allows you to match 2 distplots by bivariate data.

```python
sns.jointplot(x='total_bill',y='tip',data=tips) # Scatter plotting, HERE,pearsonr is correlation 
                                                # co-effient and the p-value for testing the correlation.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #Hexagon Distribution
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # reg --> Linear regression
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') # kde --> kernel density estimation
```

### pairplot plots pairwise relationship across entire dataframe, atleast for the numerical columns and it also supports color 'hue' argument for categorical columns. pairplot do jointplot on all possible combinations of the columns.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') #Here, 'sex' columns is categorical column
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

### rugplot will be used to explain the kde plot. It will work on univariate set of observations. It draws single dash mark for each point in the distribution. Then we will plot Gaussian distribution across each dash mark and sum of each gaussian distribution will give us KDE.
#### https://en.wikipedia.org/wiki/Kernel_density_estimation
#### https://en.wikipedia.org/wiki/Normal_distribution

```python
sns.kdeplot(tips['total_bill'], ls="")
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

### KDE Plot, kernel density or density plot visualizes the distribution of data over a continuous interval or time period. The peaks of a Density Plot displays where exactly the values are concentrated over the interval

```python
sns.kdeplot(tips['total_bill'])
```
